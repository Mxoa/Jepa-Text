```python
from data.datamodule import TextDataModule
from src.modules.jepa.encoder import TransformerEncoder
from src.modules.jepa.mlm_module import MLMModule
import lightning as L
from lightning.pytorch.loggers import WandbLogger
```

```python
dm = TextDataModule(
    train_path="data/raw/train_10000.parquet",
    val_path="data/raw/val_1000.parquet",
    tokenizer_name="bert-base-uncased",
    window_size=128,
    alpha=0.3,
    batch_size=32,
)

encoder = TransformerEncoder(
    vocab_size=30522,   # bert-base-uncased
    d_model=32,
    n_heads=2,
    n_layers=2,
    d_ff=256,
    max_seq_len=128,
)

module = MLMModule(encoder=encoder, vocab_size=30522, d_model=32, lr=1e-2)
```

```python
wandb_logger = WandbLogger(
    save_dir="logs/", 
    project="mlm_baseline"
)

trainer = L.Trainer(
    max_epochs=10,
    accelerator="cpu",
    logger=wandb_logger,
    log_every_n_steps=50,
)
```

```python
trainer.fit(module, datamodule=dm)
```
